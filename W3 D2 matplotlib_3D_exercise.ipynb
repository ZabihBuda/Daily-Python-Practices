```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

* Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>
* hyperbolic paraboloid equation : z = x*x - y*y

```python
x=np.linspace(-10, 10, 100)
y=np.linspace(-10, 10, 100)




X, Y=np.meshgrid(x,y)
Z=(X**2-Y**2)
```

```python
# wireframe
fig=plt.figure(figsize=(12, 9))

ax=plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='r')
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* eliptical paraboloid : z = x*x + y*y

```python
# surface plot
fig=plt.figure(figsize=(12, 9))

Z=(X**2 + Y**2)

ax=plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, color='b')
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* Create also contour projection on the plane z=0 (HINT: use parameter 'zdir' of contour plot)
* eliptical paraboloid : z = x*x + y*y

```python
# contour plot
fig=plt.figure(figsize=(12, 9))

Z=(X**2 + Y**2)

ax=plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='jet')

ax.contour(X, Y, Z, zdir='z', cmap='jet', offset=0)

plt.show()
```

```python
#Copied from somewhere else

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure(figsize=(12, 9))
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)

# Plot the 3D surface
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)

# Plot projections of the contours for each dimension.  By choosing offsets
# that match the appropriate axes limits, the projected contours will sit on
# the 'walls' of the graph
cset = ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)

ax.set_xlim(-40, 40)
ax.set_ylim(-40, 40)
ax.set_zlim(-100, 100)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
