```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
world_json=pd.read_json('world_bank_projects.json')
world_json.head(4)
```

### 1: Find the 10 countries with most projects

```python
projects=world_json.groupby('countryname').count()
projects.head(3)
```

```python
top_countries=projects.sort_values(by='project_name', ascending=False)
top_countries['project_name'].head(10)
```

### 2: What are top 10 sectors with projects?

```python
pd.DataFrame(world_json['sector'])
```

```python
project_sector=world_json.explode('sector')
project_sector.head()
```

```python
sector_series = project_sector['sector']
sector_series
```

```python
sector_values=pd.Series([x.get('Name') for x in sector_series])
sector_values
```

```python
project_sector['sector']=sector_values
project_sector.head(5)
```

```python
top_sectors=project_sector.groupby('sector').count()
```

```python
top_ten_sector=top_sectors.sort_values(by='project_name', ascending=False)
top_ten_sector['project_name'].head(10)
```

### 3: Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
world_json.head()
```

```python
major=world_json['mjtheme_namecode']
major
```

```python
major_themes=world_json.explode('mjtheme_namecode')
major_themes.head()
```

```python
theme_series=major_themes['mjtheme_namecode']
theme_series
```

```python
theme_names=pd.Series([x.get('name') for x in theme_series])
theme_names
```

```python
theme_numbers=pd.Series([x.get('code') for x in theme_series])
theme_numbers
```

```python
theme=pd.DataFrame()
theme['Numbers']=theme_numbers
theme['Themes']=theme_names
theme.head(10)
```

```python
theme['Themes'].replace('', np.nan, inplace=True)
theme.dropna(subset=['Themes'], inplace=True)
```

```python
theme_dict=theme.drop_duplicates()
```

```python
theme_dict = theme_dict.set_index('Numbers')['Themes'].to_dict()
theme_dict
```

```python
major_themes['Theme_Num']=theme_numbers
major_themes['Theme_Name']=major_themes['Theme_Num'].map(theme_dict)
major_themes.head()
```

```python
top_themes=major_themes.groupby('Theme_Name').count()
top_themes
```

```python
top_ten_themes=top_themes.sort_values(by=['project_name'], ascending=False)
top_ten_themes['project_name'].head(10)
```
