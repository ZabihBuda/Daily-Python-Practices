```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
%matplotlib inline
import pandas as pd
```

```python
#Line Plot
x=np.linspace(0, 100, 200)
y=np.sin(x)

fig, ax=plt.subplots(figsize=(10, 6))

ax.plot(x, y, 'red')
plt.show()
```

```python
# Scatter plot
# 200 random values from the interval <0,10>
x=10*np.random.rand(200, 1)

y=15*np.random.rand(200, 1)

fig, ax=plt.subplots(figsize=(10, 6))

ax.scatter(x, y)
plt.show()
```

```python
#Histogram
#200 random values from the interval <0,10>
y=15*np.random.rand(200, 1)

fig, ax=plt.subplots(figsize=(10, 6))

ax.hist(y, bins=20)
plt.show()
```

```python
#Graphs on Common Axes

# 200 values from the interval <0,100>, equidistantly divided
x=np.linspace(0, 100, 200)

y1=np.sin(x)
y2=(np.sin(x))*(np.cos(x))

fig, ax=plt.subplots(figsize=(10, 6))

ax.plot(x, y1, 'red')

ax.plot(x, y2, 'blue')
plt.show()
```

```python
# Subplots

# create the figure and axes, set the figure size
fig, axes=plt.subplots(nrows=2, ncols=1, figsize=(10, 6))

#the first plot
axes[0].plot(x, y1, 'red')

#the first title
axes[0].set_title('sin(x)')

#the second plot
axes[1].plot(x, y2, 'blue')

#the second plot title
axes[1].set_title('sin(x) * (cos(x))')

# automatically adjust the subplot parameters to give specified padding
fig.tight_layout()
plt.show()
```

```python
#Legends

import pandas as pd
from sklearn import datasets
```

```python
#load the dataset
iris=datasets.load_iris()

#create the dataframe
iris_df=pd.DataFrame(iris.data, columns=iris.feature_names)

#create a target
iris_df['target']=iris.target

#map target values to target names
iris_df['target_name']=iris_df.target.map({
    0:'setosa',
    1:'versicolor', 
    2:'virginica'
})

```

```python
iris_df.head()
```

```python
#Scatter Plot

#Now, let's plot a scatter plot of sepal length and sepal width for each type of Iris 
#flower (sepal is the part of a flower which is usually green):

# Iris setosa
setosa = iris_df[iris_df.target_name == 'setosa']

# Iris versicolor
versicolor = iris_df[iris_df.target_name == 'versicolor']

# Iris virginica
virginica = iris_df[iris_df.target_name == 'virginica']


# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize=(10,6))


# Scatter plot for each of them
ax.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], marker='o', color='red', label='setosa')

ax.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], marker='o', color='green', label='versicolor')

ax.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'], marker='o', color='blue', label='virginica')

#set the legend
ax.legend(loc='upper right')

#title
ax.set_title('Iris Flowers')

#x-label
ax.set_xlabel('sepal length (cm)')

#y-label
ax.set_ylabel('sepal width (cm)')

plt.show()
```

```python
# Annotations

# create the figure and axes, set the figure size
fig, ax = plt.subplots(figsize=(10,6))


# Scatter plot for each of them
ax.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], marker='o', color='red', label='setosa')

ax.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], marker='o', color='green', label='versicolor')

ax.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'], marker='o', color='blue', label='virginica')

#Setting Annotations
ax.annotate('setosa', xy=(5.0, 3.5), xytext=(4.25, 4.0), arrowprops={'color':'red'})
ax.annotate('versicolor', xy=(7.2, 3.6), xytext=(6.5, 4.0), arrowprops={'color':'red'})
ax.annotate('virginica', xy=(5.05, 1.95), xytext=(5.5, 1.75), arrowprops={'color':'red'})

#set the legend
ax.legend(loc='upper right')

#title
ax.set_title('Iris Flowers')

#x-label
ax.set_xlabel('sepal length (cm)')

#y-label
ax.set_ylabel('sepal width (cm)')

#set x-rang
ax.set_xlim(4, 8)

#set y-range
ax.set_ylim(1.5, 4.7)
plt.show()
```
