```python
# 200 values from the interval <0,100>, equidistantly divided
x=np.linspace(0, 100, 200)

#sin(x) values
y1=np.sin(x)

#sin(x)*cos(x) values
y2=(np.sin(x))*(np.cos(x))

# a line plot of sing(x), red lin
plt.plot(x, y1, 'red')

#a line plot of sin(x)*cos(x), blue line
plt.plot(x, y2, 'blue')

plt.show()
```

```python
# the first figure
plt.subplot(2, 1, 1)
plt.plot(x, y1, 'red')
plt.title('sin(x)')

#second figure
plt.subplot(2, 1, 2)
plt.plot(x, y2, 'blue')
plt.title('(sin(x))*(cos(x))')

#automatically adjust the subplot parameters to give a specified padding
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import sklearn
from sklearn import datasets
```

```python
#Loading iris dataset
iris=datasets.load_iris()

#create dataframe
iris_df=pd.DataFrame(iris.data, columns=iris.feature_names)

#create target
iris_df['target']=iris.target

#map the target values to the target names
iris_df['target_name']=iris_df.target.map({
    0:'setosa',
    1:'versicolor',
    2:'virginica'
})
```

```python
iris_df.head(5)
```

```python
#Now, let's plot a scatter plot of sepal length and sepal width for each type of the Iris flower.
setosa=iris_df[iris_df.target_name=='setosa']

virginica = iris_df[iris_df.target_name == 'virginica']
versicolor=iris_df[iris_df.target_name=='versicolor']

#ploting
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], marker ='o', color = 'red', label = 'setosa')
plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], marker='o', color='green', label='versicolor')
plt.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'], marker ='o', color = 'blue', label = 'virginica')

#legend location
plt.legend(loc='upper right')

plt.title('Iris flowers')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
#Now, let's plot a scatter plot of sepal length and sepal width for each type of the Iris flower with Annotations.
setosa=iris_df[iris_df.target_name=='setosa']

virginica = iris_df[iris_df.target_name == 'virginica']
versicolor=iris_df[iris_df.target_name=='versicolor']

#ploting
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], marker ='o', color = 'red', label = 'setosa')
plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], marker='o', color='green', label='versicolor')
plt.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'], marker ='o', color = 'blue', label = 'virginica')

#Adding annotations
plt.annotate('setosa', xy =(5.0,3.5),
             xytext = (4.25,4.0), arrowprops={'color':'red'})
plt.annotate('versicolor', xy =(7.2,3.6),
             xytext = (6.5,4.0), arrowprops={'color':'red'})
plt.annotate('virginica', xy =(5.05,1.95),
             xytext = (5.5,1.75), arrowprops={'color':'red'})

#legend location
plt.legend(loc='upper right')

plt.title('Iris flowers')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.ylim(1.5, 4.7)
plt.show()
```
