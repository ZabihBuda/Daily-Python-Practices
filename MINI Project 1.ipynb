### Task 2

What are the different modes of transport which are operated by Transfer for London? How many of modes do they have?
Print the list with different modes of transport, plus their count. 

Example output:
[bus, cable-car,.....]
Number of different modes of transport is: xyz

We need to search the documentation for correct request.

```python
#Sending request to receive information about different modes of transportation
url_mode='https://api.tfl.gov.uk/Journey/Meta/Modes'
modes=requests.get(url_mode)
```

```python
#Checking the status of the request
modes.status_code
```

```python
# Looking for the modenames and counting it.
tfl_modes = [names.get('modeName') for names in json]

print(f'The number of modes in London Transportation is:', len(tfl_modes))

print('\n',tfl_modes)
```

### Task 5
How many station has victoria line?

```python

url_station='https://api.tfl.gov.uk/Line/victoria/StopPoints'
url=requests.get(url_station)
if url.status_code==200:
    print('Success')
else:
    print('Not Found')

vict_stations=url.json()
print(f'Number of Bus Station in Victoria: {len(vict_stations)}')
for i in range(len(vict_stations)):
    stations=(vict_stations[i])
    name=stations['commonName']
    print(f'{i+1}: {name} line')
```

### Task 6

Plan the journey from Heathrow Airport to Tower Bridge using Bus and Tube? Which way is faster? 
Example output:

Planned duration:

Bus: x minutes

Tube: y minutes

We need to search the documentation for correct requests and parameters we need.

```python
url="https://api.tfl.gov.uk/Journey/JourneyResults/51.4700%2C%20-0.4543/to/51.5055%2C%20-0.0754?mode=Bus"

url_bus=requests.get(url)

bus_json=url_bus.json()
duration_by_bus=bus_json['journeys'][0]['duration']
print(f'Distance from Heathrow Airport to Bridge Tower is {duration_by_bus} minutes by Bus.')
```

```python
url="https://api.tfl.gov.uk/Journey/JourneyResults/51.4700%2C%20-0.4543/to/51.5055%2C%20-0.0754?mode=tube"

url_tube=requests.get(url)

tube_json=url_tube.json()
duration_by_tube=tube_json['journeys'][1]['duration']
print(f'Distance from Heathrow Airport to Bridge Tower is {duration_by_tube} minutes by Tube.')
```
